```python
from bs4 import BeautifulSoup
import pandas as pd

# 1. Load your SOAP HTML output (replace 'soap_output.html' with your file)
with open("soap_output.html", "r", encoding="utf-8") as f:
    html_content = f.read()

# 2. Parse HTML
soup = BeautifulSoup(html_content, "html.parser")

# 3. Extract table rows
rows = []
for tr in soup.find_all("tr")[1:]:  # skip header
    cols = tr.find_all("td")
    if len(cols) == 2:
        invoice_id = cols[0].get_text(strip=True)
        amount = cols[1].get_text(strip=True)
        rows.append([invoice_id, amount])

# 4. Create DataFrame
df = pd.DataFrame(rows, columns=["INVOICE_ID", "INVOICE_AMOUNT"])

# 5. Ensure INVOICE_ID is treated as string (avoids scientific notation in Excel)
df["INVOICE_ID"] = df["INVOICE_ID"].astype(str)
df["INVOICE_AMOUNT"] = pd.to_numeric(df["INVOICE_AMOUNT"], errors="coerce")

# 6. Export to Excel
df.to_excel("clean_invoices.xlsx", index=False)

print("✅ Excel file generated: clean_invoices.xlsx")
```
