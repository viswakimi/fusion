```python
import xml.etree.ElementTree as ET
import pandas as pd

# Sample XML
xml_data = """<DATA_DS>
<G_1><INVOICE_NUM>3000001</INVOICE_NUM><INVOICE_AMOUNT>50</INVOICE_AMOUNT><INVOICE_DATE>2018-01-30</INVOICE_DATE><VENDOR_NAME>SVARAPPS</VENDOR_NAME></G_1>
<G_1><INVOICE_NUM>3000002</INVOICE_NUM><INVOICE_AMOUNT>100</INVOICE_AMOUNT><INVOICE_DATE>2018-01-31</INVOICE_DATE><VENDOR_NAME>SVARAPPS</VENDOR_NAME></G_1>
</DATA_DS>"""

# Parse XML
root = ET.fromstring(xml_data)
records = [{child.tag: f"<<{child.text}>>" for child in g1} for g1 in root.findall('G_1')]

# Create DataFrame
df = pd.DataFrame(records)

# Save as Excel (regular workbook)
df.to_excel("Invoice_Placeholder.xlsx", index=False)
print("Excel with placeholders generated.")
```
